```python
import pandas as pd 
```

```python
matches = pd.read_csv("full.csv", index_col = 0)
```

```python
matches.head()
```

```python
matches.shape
```

```python
temp = matches ['hTeam']
matches['hTeam'] = matches['aTeam']
matches['aTeam'] = temp
```

```python
matches.head()
```

```python
odds = matches['hOdds']
nOdds = []
for odd in odds:
    split = odd.split("/")
    nOdds.append(round(int(split[0]) / int(split[1]),2))

matches['nOdds'] = nOdds
```

```python
matches['FdateTime'] = pd.to_datetime(matches['dateTime'])
```
