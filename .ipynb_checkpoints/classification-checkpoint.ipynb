```python
import pandas as pd 
import numpy as np
from sklearn.metrics import accuracy_score, precision_score
from sklearn.ensemble import RandomForestClassifier
```

```python
matches = pd.read_csv("full.csv")
```

```python
matches.columns
```

```python
matches.shape
```

```python
temp = matches ['hTeam']
matches['hTeam'] = matches['aTeam']
matches['aTeam'] = temp
```

```python
odds = matches['hOdds']
nOdds = []
for odd in odds:
    split = odd.split("/")
    nOdds.append(round(int(split[0]) / int(split[1]),2))

matches['nOdds'] = nOdds
```

```python
matches['FdateTime'] = pd.to_datetime(matches['dateTime'])
```

```python
matches['hTeamCode'] = matches['hTeam'].astype('category').cat.codes
```

```python
matches['aTeamCode'] = matches['aTeam'].astype('category').cat.codes
```

```python
matches['dateTime'] = pd.to_datetime(matches['dateTime'])
matches['dateCode'] = matches['dateTime'].dt.strftime('%Y%m%d').astype(int)
```

```python
matches['timeCode'] = matches['dateTime'].dt.strftime('%H%m%s').astype(int)
```

```python
matches.drop('FdateTime', axis = "columns")
```

```python
rf = RandomForestClassifier (n_estimators = 50, min_samples_split = 10, random_state = 1)
```

```python
matches['betFail'] = np.where(
    ((matches['hWin'] == 0) & (matches['nOdds'] < 1)) |
    ((matches['hWin'] == 1) & (matches['nOdds'] > 1)),
    0,  
    1   
)
```

```python
matches
```

```python
predictors = ['nOdds','hTeamCode','aTeamCode','dateCode','timeCode','hWin']
X = matches[predictors]
y = matches['betFail']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split (X,y,random_state = 0)
```

```python
rf.fit (X_train,y_train)
```

```python
preds = rf.predict(X_test)
```

```python
acc = accuracy_score(y_test,preds)
```

```python
acc
```

```python
precision_score(y_test,preds)
```

```python
def rolling_averages (group,cols,new_cols):
    group = group.sort_values('timeCode')
    rolling_stats = group[cols].rolling(3,closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group

def weighted_form(arr):
    weights = np.arange(1, len(arr) + 1)
    print(weight)
    return np.sum(arr * weights) / np.sum(weights)

print (group['hWin'].rolling(3,closed='left').apply(weighted_form, raw=True))
```

```python
cols = 'hBall possession,aBall possession,hExpected goals,aExpected goals,hBig chances,aBig chances,hTotal shots,aTotal shots,hGoalkeeper saves,aGoalkeeper saves,hCorner kicks,aCorner kicks,hFouls,aFouls,hPasses,aPasses,hTackles,aTackles,hFree kicks,aFree kicks,hYellow cards,aYellow cards,aRed Cards,hRed Cards'.split(",")
```

```python
new_cols = [f"{c}_rolling" for c in cols]
rolling_averages (group,cols,new_cols)
```

```python
matches['hBall possession'] = matches['hBall possession'].str.rstrip('%').astype(float) / 100
matches['aBall possession'] = matches['aBall possession'].str.rstrip('%').astype(float) / 100
```

```python
matches_rolling = matches.groupby('hTeam').apply (lambda x: rolling_averages(x,cols,new_cols))
```

```python
def make_prediction (data,predictors):
    rf = RandomForestClassifier (n_estimators = 50, min_samples_split = 10, random_state = 1)
    X = data[predictors]
    y = data['betFail']
    X_train,X_test,y_train,y_test = train_test_split (X,y,random_state = 0)
    rf.fit (X_train,y_train)
    preds = rf.predict(X_test)
    acc = accuracy_score (y_test,preds)
    precision = precision_score (y_test,preds)
    print (acc,precision)
```

```python
new_predictors = predictors + new_cols
print (new_predictors)
make_prediction (matches_rolling,new_predictors)
```
